```python
import pandas as pd
import seaborn as sns
import pygwalker as pyg
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D

def drawLineGraph(df, colName, targetName):
    plt.figure(figsize=(20,6))
    sns.lineplot(data=df, y=df[colName], x=df.index, hue=targetName)
    plt.show()
 
# 데이터를 DataFrame에 로드
data = pd.read_csv('./remove_outlier_dataset.csv', encoding='utf-8', index_col=0)
data["registration_time"] = pd.to_datetime(data["registration_time"])
data["mold_code"] = data["mold_code"].astype(str)
data["passorfail"] = data["passorfail"].astype(str)
# DataFrame으로 Graphic Walker 호출
# [data["passorfail"]=="1.0"]
gwalker = pyg.walk(data)
```

```python
fig = plt.figure(figsize=(10,9))
ax = Axes3D(fig)
fig.add_axes(ax)

ax.scatter(data=data, xs="cast_pressure", ys="biscuit_thickness", zs="lower_mold_temp2", c="passorfail" ,cmap="inferno", s=10)
```

```python
from sklearn.preprocessing import RobustScaler
feature_names = ['count','cast_pressure', 'biscuit_thickness','upper_mold_temp1', 'upper_mold_temp2', 'lower_mold_temp1', 'lower_mold_temp2',
                    'sleeve_temperature']
# 데이터를 DataFrame에 로드
df = pd.read_csv('./remove_outlier_dataset.csv', encoding='utf-8', index_col=0)
df["registration_time"] = pd.to_datetime(df["registration_time"])
df.set_index(keys="registration_time",inplace=True)
df["mold_code"] = df["mold_code"].astype(str)
df["passorfail"] = df["passorfail"].astype(int).astype(str)

featureDf = df[feature_names]

scaler = RobustScaler()
scaler.fit_transform(featureDf)
scaled_df = pd.DataFrame(data=scaler.fit_transform(featureDf), columns=feature_names, index=df.index)
scaled_df["passorfail"] = df["passorfail"]
```

```python

sampleData = scaled_df[(scaled_df.index>="2019-01-01")&(scaled_df.index<="2019-01-03")]

plt.figure(figsize=(20,6))
sns.lineplot(data=sampleData, y=sampleData["cast_pressure"], x=sampleData.index, hue="passorfail")
sns.lineplot(data=sampleData, y=sampleData["biscuit_thickness"], x=sampleData.index, hue="passorfail")
# sns.scatterplot(data=sampleData, y=sampleData["upper_mold_temp2"], x=sampleData.index, hue="passorfail")
# sns.scatterplot(data=sampleData, y=sampleData["sleeve_temperature"], x=sampleData.index, hue="passorfail")
plt.show()
```
